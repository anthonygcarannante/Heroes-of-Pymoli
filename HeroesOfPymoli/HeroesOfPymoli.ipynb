```python
import pandas as pd
```

```python
pymoli_file = "../Instructions/HeroesOfPymoli/Resources/purchase_data.csv"
pymoli_df = pd.read_csv(pymoli_file)
pymoli_df.head()
```

```python
### PURCHASING ANALYSIS (TOTAL)
unique_players = len(pymoli_df['SN'].unique())

unique_items = pymoli_df['Item ID'].value_counts()
average_item_price = pymoli_df['Price'].mean()
total_purchases = pymoli_df['Purchase ID'].count()
total_revenue = pymoli_df['Price'].sum()

summary_table1 = pd.DataFrame({'Unique Items': len(unique_items), 
                               'Average Item Price': average_item_price,
                               'Number of Purchases': total_purchases, 
                               'Total Revenue': total_revenue}, index = [0])

summary_table1['Total Revenue'] = summary_table1['Total Revenue'].map('${:,.2f}'.format)
summary_table1['Average Item Price'] = summary_table1['Average Item Price'].map('${:,.2f}'.format)

summary_table1
```

```python
### GENDER DEMOGRAPHICS
male_count = pymoli_df.loc[pymoli_df['Gender'] == 'Male',:]
female_count = pymoli_df.loc[pymoli_df['Gender'] == 'Female',:]
other_count = pymoli_df.loc[pymoli_df['Gender'] == 'Other / Non-Disclosed',:]

male_count_clean = male_count['SN'].value_counts()
female_count_clean = female_count['SN'].value_counts()
other_count_clean = other_count['SN'].value_counts()
total_count = len(male_count_clean) + len(female_count_clean) + len(other_count_clean)

gender_df = pd.DataFrame({'Gender':['Male','Female','Other/Not_Specified'], 
                          'Value': [len(male_count_clean),len(female_count_clean),len(other_count_clean)]
                         })
gender_df['Percentage of Players'] = ["{:.2%}".format(len(male_count_clean)/total_count), 
                                      "{:.2%}".format(len(female_count_clean)/total_count), 
                                      "{:.2%}".format(len(other_count_clean)/total_count)]
gender_df
```
